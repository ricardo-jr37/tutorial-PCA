```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 15})
plt.rcParams['figure.figsize'] = 18, 6
pd.set_option('display.max_colwidth', 50)
```

```python
"""
https://stackoverflow.com/questions/16024677/generate-correlated-data-in-python-3-3
"""
#gerando uma lista correlacionada
linalg = np.linalg
num_samples = 100
num_variables = 2
cov = [[1, 0.9], 
       [0.9, 1]]

L = linalg.cholesky(cov)
uncorrelated = np.random.standard_normal(size=(num_samples,num_variables))
mean = [75, 75]
#correlated = np.dot(L, uncorrelated.T) + np.array(mean).reshape(2, 1)
correlated = np.dot(L, uncorrelated.T)
xy = correlated.T
x = xy[:, 0].copy()
y = xy[:, 1].copy()
```

```python
import numpy.random as rd
v = rd.normal(75, 5, (100,1))
dataset = correlated.T.copy()
dataset[:,0] += v[0]
dataset[:,1] += v[0]
x = dataset[:, 0].copy()
y = dataset[:, 1].copy()
```

### O que estamos fazendo quando fazemos o produto interno entre as variáveis?

```python
plt.scatter(scaled[:, 0], scaled[:, 1], c='green')
plt.plot([-3,3], [-3,3], c='red')
plt.show()
```

```python
autovalores, autovetores = np.linalg.eig(matrizcov)
```

```python
print("Auto-valores:\n {}\nAuto-vetores:\n {}\n".format(autovalores, autovetores))
```

```python
pares_de_auto=[
    (
        np.abs(autovalores[i]),
        autovetores[:, i]
    )for i in range(len(autovalores))
]
pares_de_auto.sort()
pares_de_auto.reverse()
pares_de_auto
```

```python
p1 = pares_de_auto[0][1][0]/pares_de_auto[1][1][0]
p2 = pares_de_auto[1][1][1]/pares_de_auto[0][1][1]
```

## A soma dos autovalores é igual a soma das variâncias dos nossos atributos

```python
np.sum(autovalores)
np.var(scaled[:,1])+np.var(scaled[:,0])
```

```python
#autovetores.sort()
scores = np.dot(scaled, autovetores)
```

```python
axes = plt.gca()
plt.scatter(scaled[:, 0], scaled[:, 1], c='green')
x_vals = np.array(axes.get_xlim())
y_vals = p1 * x_vals
plt.plot(x_vals, y_vals, '--', color = 'r', label="PC 1")
x_vals = np.array(axes.get_xlim())
y_vals = p2 * x_vals
plt.plot(x_vals, y_vals, '--', color = 'b', label="PC 2")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.tight_layout(pad=1)
plt.savefig("output.png")
plt.show()
```

```python
plt.figure(figsize = (12.24, 3.56))
plt.scatter(scores[:,0], scores[:,1], color='green')
plt.axhline(lw=1, color ='r', label= 'PC 1')
plt.axvline(lw=1, color ='b', label= 'PC 2') 
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.tight_layout(pad=1)
plt.show()
```

```python
dataset = pd.read_csv('wine.data', sep = ',',names=range(1, 15, 1))
x = dataset.drop(columns=[1]).copy()
y = dataset[1].copy()
display(dataset)
```

```python
import seaborn as sns
sns.pairplot(dataset, hue=1)
```

```python
scaled = dataset.to_numpy()
x = scaled[:, 1:14].copy()
y = scaled[:, 0].copy()
```

```python
for i in range(1, scaled.shape[1]):
    min = np.amin(scaled[:,i]) 
    max = np.amax(scaled[:,i])
    print("Atributo {}:".format(i+1))
    print("Minimum =", min) 
    print("Maximum = {}\n".format(max))
    #print('\n')
```

```python
for i in range(0, x.shape[1]):
    x[:, i] = (x[:, i] - np.mean(x,  axis=0)[i])/np.std(x,  axis=0)[i]
```

```python
#Valores bem mais próximos entre si, depois de escalados
for i in range(0, x.shape[1]):
    min = np.amin(x[:,i]) 
    max = np.amax(x[:,i])
    print("Atributo {}:".format(i+2))
    print("Minimum =", min) 
    print("Maximum =", max)
    print('\n')
```

```python
matrizcov = np.cov(x, rowvar=0)
autovalores, autovetores = np.linalg.eig(matrizcov)
print("Auto-valores:\n {}\nAuto-vetores:\n {}\n".format(autovalores, autovetores))
```

```python
pares_de_auto=[
    (
        np.abs(autovalores[i]),
        autovetores[:, i]
    )for i in range(len(autovalores))
]
pares_de_auto.sort()
pares_de_auto.reverse()
pares_de_auto
```

```python
total = sum(autovalores)
var_exp = [
    (i/total)*100 for i in sorted(
        autovalores, reverse = True
    )
]
cum_var_exp = np.cumsum(var_exp)
```

```python
cum_var_exp
```

```python
plt.scatter(range(1, len(cum_var_exp)+1), cum_var_exp)
plt.show()
```
